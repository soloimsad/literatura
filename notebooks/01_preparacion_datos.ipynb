# 01 - Preparacion de datos local

Audita los datasets locales y genera los YAML/formatos necesarios para los modelos del proyecto.
El flujo asume que todo se ejecuta en este equipo, desde la raiz del repositorio.

```python
from pathlib import Path
import os
import subprocess
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

print("PROJECT_ROOT =", PROJECT_ROOT)
```

```python
from pathlib import Path

paths = {
    "dataset_pieza_dental": PROJECT_ROOT / "data" / "teeth_segmentation",
    "dataset_tratamiento": PROJECT_ROOT / "data" / "dental-disease-panoramic-detection-dataset",
    "datos_preparados_pieza": PROJECT_ROOT / "data" / "_prepared_teeth_yolo_seg",
    "resultados": PROJECT_ROOT / "results",
}

for name, path in paths.items():
    print(f"{name}: {path} | existe={path.exists()}")
```

```python
# Auditoria y preparacion del dataset de pieza dental.
env = os.environ.copy()
env.update({
    "ANALYZE_ONLY": "1",
    "TRAIN_DATASET": "teeth",
    "PROJECT_DIR": str(PROJECT_ROOT / "results" / "data_preparation" / "tooth"),
})

subprocess.run(
    [sys.executable, "src/dental_xray_pipeline.py"],
    cwd=PROJECT_ROOT,
    env=env,
    check=True,
)
```

```python
# Auditoria y preparacion del dataset de posibles tratamientos/hallazgos.
env = os.environ.copy()
env.update({
    "ANALYZE_ONLY": "1",
    "TRAIN_DATASET": "disease",
    "PROJECT_DIR": str(PROJECT_ROOT / "results" / "data_preparation" / "treatment"),
})

subprocess.run(
    [sys.executable, "src/dental_xray_pipeline.py"],
    cwd=PROJECT_ROOT,
    env=env,
    check=True,
)
```

```python
# Archivos generados mas importantes.
for path in [
    PROJECT_ROOT / "results" / "data_preparation" / "tooth" / "DATASET_ANALYSIS.md",
    PROJECT_ROOT / "results" / "data_preparation" / "tooth" / "preparation_summary.json",
    PROJECT_ROOT / "results" / "data_preparation" / "treatment" / "DATASET_ANALYSIS.md",
    PROJECT_ROOT / "results" / "data_preparation" / "treatment" / "prepared_yolo_seg" / "data.yaml",
]:
    print(path, "| existe=", path.exists())
```
