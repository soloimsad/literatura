```python
from pathlib import Path
import os
import subprocess
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

print("PROJECT_ROOT =", PROJECT_ROOT)
```

```python
from model_registry import get_model_spec, registry_rows

for row in registry_rows():
    print(row)
```

```python
IMAGE_PATH = PROJECT_ROOT / "data" / "samples" / "sample-19.jpg"
OUTPUT_DIR = PROJECT_ROOT / "results" / "presentation_demo"

role = "both" if get_model_spec("treatment").exists else "tooth"
print("Rol ejecutado:", role)

cmd = [
    sys.executable,
    "src/predict_dental_xray.py",
    "--model-role", role,
    "--image", str(IMAGE_PATH),
    "--output-dir", str(OUTPUT_DIR),
    "--device", "cpu",
    "--imgsz", "640",
    "--conf", "0.25",
]
subprocess.run(cmd, cwd=PROJECT_ROOT, check=True)
```

```python
from IPython.display import Image, display

for image in sorted(OUTPUT_DIR.rglob("*_labeled.jpg")):
    print(image)
    display(Image(filename=str(image)))

for report in sorted(OUTPUT_DIR.rglob("*_report.csv")):
    print("\n", report)
    print("\n".join(report.read_text(encoding="utf-8").splitlines()[:8]))
```
